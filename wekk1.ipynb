```python
expenses = [
    {"category": "Food", "amount": 320},
    {"category": "Transport", "amount": 150},
    {"category": "Food", "amount": 210},
    {"category": "Entertainment", "amount": 800},
    {"category": "Transport", "amount": 95},
    {"category": "Food", "amount": 430},
]
category = set()
spent = 0
food_money = 0
transport_money = 0
entertainment_money = 0

for i in expenses:
    category.add(i["category"])
    if i["amount"] > 500:
        continue
    spent += i["amount"]
    if i["category"] == "Food":
        food_money += i["amount"]
    if i["category"] == "Transport":
        transport_money += i["amount"]
    if i["category"] == "Entertainment":
        entertainment_money += i["amount"]


cats = {

}

cats["food"] = food_money
cats["transport"] = transport_money
cats["Entertainment"] = entertainment_money


print(category)
print(f"Total Spent: {spent}")
for i, j in cats.items():
    print(f"{i}: {j}")
```
